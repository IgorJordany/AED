# Tratando Dados Ausentes

```python
import pandas as pd
professores2_df = pd.read_excel('./datas/dados_2.xlsx', index_col=0)
```

```python
professores2_df
```

Removendo linhas com dados ausentes

```python
professores2_df_no_missing = professores2_df.dropna()
professores2_df_no_missing
```

Muitos dados foram removidos

Removendo linhas com a maioria dos dados ausentes

```python
professores2_df
```

```python
professores2_df_limiar_missing = professores2_df.dropna(thresh=5)
professores2_df_limiar_missing
```

```python
professores2_df_limiar_missing = professores2_df.dropna(thresh=6)
professores2_df_limiar_missing
```

Removendo com colunas com percentual de dados ausentes acima do parÃ¢metro definido

```python
professores2_df
```

```python
professores2_df_percentual_columns_missing = professores2_df.dropna(axis=1, thresh=int(0.6*len(professores2_df)))
professores2_df_percentual_columns_missing
```

Preenchendo dados ausentes com valores genÃ©ricos.

```python
professores2_df
```

```python
professores2_df_generic_no_missing = professores2_df.fillna(-1)
professores2_df_generic_no_missing
```

```python
professores2_df
```

```python
professores2_df_next_value = professores2_df.bfill()  #Utiliza o prÃ³ximo valor nÃ£o nulo
professores2_df_next_value
```

```python
professores2_df
```

```python
professores2_df_forward_value = professores2_df.ffill() #Utiliza o valor anterior nÃ£o nulo
professores2_df_forward_value
```

Preenchimento de dados ausentes com valores de tendÃªncias centrais.

```python
professores2_df
```

```python
idade_media = professores2_df['idade'].mean()
altura_media = professores2_df['altura'].mean()
print("Idade MÃ©dia:", idade_media, " - Altura MÃ©dia:", altura_media)
```

```python
dados_preenchimento = {'idade': idade_media, 'altura': altura_media} # DicionÃ¡rio
dados_preenchimento
```

```python
professores2_df_central_tendency=professores2_df.fillna(value=dados_preenchimento)
professores2_df_central_tendency
```

Preenchimento de dados ausentes baseado em condiÃ§Ãµes.

```python
professores2_df = pd.read_excel('./datas/dados_2.xlsx', index_col=0)
professores2_df
```

```python
professores2_df_peso_genero = professores2_df[['genero', 'peso']]
```

```python
professores2_df_peso_genero
```

```python
professores2_df_peso_genero.groupby('genero').transform(lambda x: x.fillna(x.mean()))
```

```python
professores2_df_peso_genero
```

```python
professores2_df
```

```python
professores2_df[['peso']] = professores2_df_peso_genero[['peso']]
```

```python
professores2_df
```

```python
for x in professores2_df.groupby("genero"):
    print("Imprimindo grupo")
    print(x)
    print("\n\n")
```
